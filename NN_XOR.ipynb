```python
import numpy as np
```

```python
# Веса
w = [np.random.normal(),np.random.normal(),np.random.normal(),np.random.normal(),np.random.normal(),np.random.normal()]
global w 

# Смещения
b = [np.random.normal(),np.random.normal(),np.random.normal()]
global b
```

```python
def sigmoid(x):
    # Функция активации sigmoid:: f(x) = 1 / (1 + e^(-x))
    return 1 / (1 + np.exp(-x))
 
 
def deriv_sigmoid(x):
    # Производная от sigmoid: f'(x) = f(x) * (1 - f(x))
    fx = sigmoid(x)
    return fx * (1 - fx)
 
 
def mse_loss(y_true, y_pred):
    # y_true и y_pred являются массивами numpy с одинаковой длиной
    return ((y_true - y_pred) ** 2).mean()
```

```python
def feedforward(x):
        # x является массивом numpy с двумя элементами
        h1 = sigmoid(w[0] * x[0] + w[1] * x[1] + b[0])
        h2 = sigmoid(w[2] * x[0] + w[3] * x[1] + b[1])
        o1 = sigmoid(w[4] * h1 + w[5] * h2 + b[2])
        return o1
```

```python
def train(data, all_y_trues):
        learn_rate = 0.1
        epochs = 4000 # количество поколений во всём наборе данных
 
        for epoch in range(epochs):
            for x, y_true in zip(data, all_y_trues):
                # --- Выполняем обратную связь (нам понадобятся эти значения в дальнейшем)
                sum_h1 = w[0] * x[0] + w[1] * x[1] + b[0]
                h1 = sigmoid(sum_h1)
 
                sum_h2 = w[2] * x[0] + w[3] * x[1] + b[1]
                h2 = sigmoid(sum_h2)
 
                sum_o1 = w[4] * h1 + w[5] * h2 + b[2]
                o1 = sigmoid(sum_o1)
                y_pred = o1
 
                # --- Подсчет частных производных
                # --- Наименование: d_L_d_w1 представляет "частично L / частично w1"
                d_L_d_ypred = -2 * (y_true - y_pred)
 
                # Нейрон o1
                d_ypred_d_w5 = h1 * deriv_sigmoid(sum_o1)
                d_ypred_d_w6 = h2 * deriv_sigmoid(sum_o1)
                d_ypred_d_b3 = deriv_sigmoid(sum_o1)
 
                d_ypred_d_h1 = w[4] * deriv_sigmoid(sum_o1)
                d_ypred_d_h2 = w[5] * deriv_sigmoid(sum_o1)
 
                # Нейрон h1
                d_h1_d_w1 = x[0] * deriv_sigmoid(sum_h1)
                d_h1_d_w2 = x[1] * deriv_sigmoid(sum_h1)
                d_h1_d_b1 = deriv_sigmoid(sum_h1)
 
                # Нейрон h2
                d_h2_d_w3 = x[0] * deriv_sigmoid(sum_h2)
                d_h2_d_w4 = x[1] * deriv_sigmoid(sum_h2)
                d_h2_d_b2 = deriv_sigmoid(sum_h2)
 
                # --- Обновляем вес и смещения
                # Нейрон h1
                w[0] -= learn_rate * d_L_d_ypred * d_ypred_d_h1 * d_h1_d_w1
                w[1] -= learn_rate * d_L_d_ypred * d_ypred_d_h1 * d_h1_d_w2
                b[0] -= learn_rate * d_L_d_ypred * d_ypred_d_h1 * d_h1_d_b1
 
                # Нейрон h2
                w[2] -= learn_rate * d_L_d_ypred * d_ypred_d_h2 * d_h2_d_w3
                w[3] -= learn_rate * d_L_d_ypred * d_ypred_d_h2 * d_h2_d_w4
                b[1] -= learn_rate * d_L_d_ypred * d_ypred_d_h2 * d_h2_d_b2
 
                # Нейрон o1
                w[4] -= learn_rate * d_L_d_ypred * d_ypred_d_w5
                w[5] -= learn_rate * d_L_d_ypred * d_ypred_d_w6
                b[2] -= learn_rate * d_L_d_ypred * d_ypred_d_b3
 
            # --- Подсчитываем общую потерю в конце каждой фазы
            if epoch % 10 == 0:
                y_preds = np.apply_along_axis(feedforward, 1, data)
                loss = mse_loss(all_y_trues, y_preds)
                print("Epoch %d loss: %.3f" % (epoch, loss))
 
data = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1],
])
 
all_y_trues = np.array([
    0,
    1,
    1,
    0,
])
 
# Тренируем нашу нейронную сеть!
train(data, all_y_trues)

print('\n')

for i in range(len(w)):
  print('w[', i , '] = ',  w[i]) 

print('\n')

for i in range(len(b)):
  print('b[', i , '] = ',  b[i]) 
```

```python
# Делаем предсказания
t1 = np.array([0, 0])
t2 = np.array([0, 1])
t3 = np.array([1, 0])
t4 = np.array([1, 1])

# В коментарии написан один из получившихся результатов
print("T1: %.3f" % feedforward(t1)) # T1: 0.029
print("T2: %.3f" % feedforward(t2)) # T2: 0.975
print("T3: %.3f" % feedforward(t3)) # T3: 0.975
print("T4: %.3f" % feedforward(t4)) # T4: 0.026

# w[ 0 ] =  6.077642015488218
# w[ 1 ] =  6.042741629686746
# w[ 2 ] =  4.509783487327476
# w[ 3 ] =  4.499012603304483
# w[ 4 ] =  8.792649628343577
# w[ 5 ] =  -9.354576310085942


# b[ 0 ] =  -2.655263195277418
# b[ 1 ] =  -6.915209109106616
# b[ 2 ] =  -4.0832346283975465
```
